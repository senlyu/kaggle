### Find items that test data has but are not contained in the training data

```python
data = pd.read_csv("train.csv")
```

```python
data_item = data[['item_nbr']]
```

```python
dat_item = dat[['item_nbr']]
```

```python
data_item = data_item.drop_duplicates()
```

```python
data_item.shape
```

```python
dat_item = dat_item.drop_duplicates()
```

```python
dat_item.shape
```

```python
common = data_item.merge(dat_item, on = "item_nbr")
```

```python
common = common['item_nbr'].tolist()
```

```python
data_item = data_item['item_nbr'].tolist()
```

```python
data_item
```

```python
dat_item = dat_item['item_nbr'].tolist()
```

```python
dat_item
```

```python
test_not_train = list(set(dat_item) - set(common))
```

```python
test_not_train
```

```python
df = pd.DataFrame(test_not_train)
```

```python
df.columns = ['item_nbr']
df
```

```python
df.to_csv('item_in_test_but_not_in_train.csv', index = False)
```
